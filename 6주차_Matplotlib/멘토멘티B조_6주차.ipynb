```python
N = 30
np.random.seed(0)
x = np.random.rand(N)
y1 = np.random.rand(N)
y2 = np.random.rand(N)
y3 = np.pi * (15 * np.random.rand(N))**2
plt.title("Bubble Chart")
plt.scatter(x, y1, c=y2, s=y3)
plt.show()
```

# 3. 그래프를 좀 더 가시적으로 만들 수 있는 방법들

## 그림 범위 지정
* 플롯 그림을 보면 몇몇 점들은 그림의 범위 경계선에 있어서 잘 보이지 않는 경우가 있을 수 있다. 
* 그림의 범위를 수동으로 지정하려면 `xlim` 명령과 `ylim` 명령을 사용한다.
* 이 명령들은 그림의 범위가 되는 x축, y축의 `최소값`과 `최대값`을 지정한다.

```python
plt.plot([10, 20, 30, 40], [1, 4, 9, 16],
         c="b", lw=5, ls="--", marker="o", ms=15, mec="g", mew=5, mfc="r")
plt.xlim(0, 50) # X축의 최소값을 0, 최대값을 50으로 지정
plt.ylim(-10, 30) # Y축의 최소값을 -10, 최대값을 30으로 지정
plt.show()
```

## 틱(tick) 설정
* 플롯이나 차트에서 축상의 위치 표시 지점을 `틱(tick)`이라고 하고 이 틱에 써진 숫자 혹은 글자를 `틱 라벨(tick label)`이라고 한다. 
* 틱의 위치나 틱 라벨은 Matplotlib가 자동으로 정해주지만 만약 수동으로 설정하고 싶다면 `xticks` 명령이나 `yticks` 명령을 사용한다.

```python
X = np.linspace(-np.pi, np.pi, 256)
C = np.cos(X)
plt.plot(X, C)
plt.xticks([-np.pi, -np.pi / 2, 0, np.pi / 2, np.pi])
plt.yticks([-1, 0, +1])
plt.show()
```

```python
# 틱 라벨 문자열에는 "$$" 사이에 LaTeX 수학 문자식을 넣을 수도 있다.
X = np.linspace(-np.pi, np.pi, 256)
C = np.cos(X)
plt.plot(X, C)
plt.xticks([-np.pi, -np.pi / 2, 0, np.pi / 2, np.pi],
           [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])
plt.yticks([-1, 0, 1], ["Low", "Zero", "High"])
plt.show()
```

## 범례
- 여러개의 라인 플롯을 동시에 그리는 경우에는 각 선이 무슨 자료를 표시하는지를 보여주기 위해 legend 명령으로 범례(legend)를 추가할 수 있다. 
- 범례의 위치는 자동으로 정해지지만 수동으로 설정하고 싶으면 `loc 인수`를 사용한다. 
- 인수에는 문자열 혹은 숫자가 들어가며 가능한 코드는 다음과 같다.

<img src= '범례.PNG'>


```python
# 범례 위치: upper left 
X = np.linspace(-np.pi, np.pi, 256)
C = np.cos(X)
S = np.sin(X)
plt.plot(X, C, ls="--", label="cosine")  # 이 플랏이 의미하는 것이 무엇인지 label로 지정
plt.plot(X, S, ls=":", label="sine")
plt.legend(loc=2)
plt.show()
```

```python
# 범례 위치: center
X = np.linspace(-np.pi, np.pi, 256)
C = np.cos(X)
S = np.sin(X)
plt.plot(X, C, ls="--", label="cosine")
plt.plot(X, S, ls=":", label="sine")
plt.legend(loc=10)
plt.show()
```

## X축, Y축 라벨과 타이틀

- 플롯의 x축 위치와 y축 위치에는 각각 그 데이터가 의미하는 바를 표시하기 위해 라벨(label)를 추가할 수 있다. 
- 라벨을 붙이려면 `xlabel`. `ylabel` 명령을 사용한다.
- 또 플롯의 위에는 `title` 명령으로 제목(title)을 붙일 수 있다.

```python
X = np.linspace(-np.pi, np.pi, 256)
C = np.cos(X)
S = np.sin(X)
plt.plot(X, C, ls="--", label="cosine")  
plt.plot(X, S, ls=":", label="sine")
plt.legend(loc=2) # 범례 위치 지정
plt.xlabel("time") # X축 이름 지정
plt.ylabel("amplitude") # Y축 이름 지정
plt.title("Cosine & Sine plot") # 제목 지정
plt.show()
```

## 어노테이션(annotation)
- `annotate` 명령을 사용하면 그림 내에 화살표를 포함한 부가 설명 문자 열을 넣을 수 있다
- `plt.annotate('문자열', xy, xytext, arrowprops)`
- 문자열은 어노테이션에서 나타낼 문자열
- xy는 화살표가 가르키는 점의 위치
- xytext는 문자열이 출력될 위치
- arrowprops는 화살표의 속성으로 칼라 등을 정의한다.



```python

x = np.arange(1,10)
y = x*5

plt.plot(x,y)
plt.annotate('annotate',xy=(2,10),xytext=(5,20),arrowprops={'color':'green'})
plt.show()
```

## 2개의 y축 한 figure에서 사용하기
여러가지 플롯을 하나의 Axes 객체에 표시할 때 y값의 크기가 달라서 표시하기 힘든 경우가 있다. 이 때는 다음처럼 `twinx` 명령으로 대해 복수의 y 축을 가진 플롯을 만들수도 있다. twinx 명령은 x 축을 공유하는 새로운 Axes 객체를 만든다.

```python
fig, ax0 = plt.subplots()
ax1 = ax0.twinx()
ax0.plot([10, 5, 2, 9, 7], 'r-', label="y0")
ax0.set_ylabel("y0")
ax0.grid(False)
ax1.plot([100, 200, 220, 180, 120], 'g:', label="y1")
ax1.set_ylabel("y1")
ax1.grid(False)
ax0.set_xlabel("x")
plt.show()
```

## subplot()
`plt.subplot(행, 열, 몇 번째 그래프)`
- 아래의 순서에 따라 번호가 매겨지므로 3번째 인수에 이 숫자를 참고하여 넣으면 된다.
<img src = '표.png'>

```python
# 예시1
x = np.arange(1,10)
y1 = x*5
y2 = x*1
y3 = x*0.3
y4 = x*0.2

plt.subplot(2,1,1)
plt.plot(x,y1)
plt.subplot(2,1,2)
plt.plot(x,y2)
plt.show()


```

```python
# 예시2
x = np.arange(1,10)
y1 = x*5
y2 = x*1
y3 = x*0.3
y4 = x*0.2

plt.subplot(1,2,1)
plt.plot(x,y1)
plt.subplot(1,2,2)
plt.plot(x,y2)
plt.show()

```

- 만약 2x2 형태의 네 개의 플롯이라면 다음과 같이 그린다.
- 이 때 subplot 의 인수는 (2,2,1)를 줄여서 221 라는 하나의 숫자로 표시할 수도 있다. 
- Axes의 위치는 위에서 부터 아래로, 왼쪽에서 오른쪽으로 카운트한다.

```python
x = np.arange(1,10)
y1 = x*5
y2 = x*1
y3 = x*0.3
y4 = x*0.2

plt.subplot(2,2,1)
plt.plot(x,y1)
plt.subplot(2,2,2)
plt.plot(x,y2)
plt.subplot(2,2,3)
plt.plot(x,y3)
plt.subplot(2,2,4)
plt.plot(x,y4)
plt.show()


```

## subplots()
- subplots 명령으로 복수의 Axes 객체를 동시에 생성할 수도 있다. 이때는 2차원 ndarray 형태로 Axes 객체가 반환된다.
- 즉,  한 번에 여러개의 그래프를 그리려면 `plt.subplots(행, 열)` 를 이용하면 된다.
- plt.subplots(행, 열) 은 2개의 변수를 반환하는데,
- 하나는 전체 액자 figure에 대한 변수고,다른 하나는 액자 내 여러개의 액자에 대한 리스트 axs다.


```python
fig, ax = plt.subplots(1, 3) #1행 3열
plt.show()
```

```python
# 1행 3열로 액자를 그린다.
fig, ax = plt.subplots(1, 3)

#각 액자들에 그래프를 그린다.
ax[0].plot([1,2,3],[1,2,3])
ax[1].bar([1,2,3],[1,2,3]) 
ax[2].hist([1,2,3], [1,2,3])
plt.show()
```

```python
# 좀 더 이쁘게 만들어 보자!
# figure 크기를 조절할 수 있다. figsize(가로길이, 세로길이)
# 가로20, 세로10인 큰 액자 안에, 1행 3열로 액자들을 그린다. 

fig, ax = plt.subplots(1, 3, figsize=(20, 5))

# 각 액자들에 그래프를 그린다. 
ax[0].plot([1,2,3],[1,2,3]) 
ax[1].bar([1,2,3],[1,2,3]) 
ax[2].hist([1,2,3], [1,2,3])
plt.show()
```

```python
# 이번엔 색깔을 바꿔볼까요?
# 가로20, 세로10인 큰 액자 안에, 1행 3열로 액자들을 그린다. 
fig, ax = plt.subplots(1, 3, figsize=(20, 5))

# 각 액자들에 그래프를 그린다. 
ax[0].plot([1,2,3],[1,2,3], 'r--') # red, 점선
ax[1].bar([1,2,3],[1,2,3], color= 'y') # yellow
ax[2].hist([1,2,3], [1,2,3], color= 'c') #cyan
plt.show()
```

```python
# 이번엔 축제목과 그래프 제목까지!!!!
# 가로20, 세로10인 큰 액자 안에, 1행 3열로 액자들을 그린다. 
fig, ax = plt.subplots(1, 3, figsize=(20, 5))

# 각 액자들에 그래프를 그린다. 
ax[0].plot([1,2,3],[1,2,3], 'r--') 
ax[0].set_title("0 title") 
ax[0].set_xlabel("0 X") 
ax[0].set_ylabel("0 Y")

ax[1].bar([1,2,3],[1,2,3], color= 'y') 
ax[1].set_title("1 title") 
ax[1].set_xlabel("1 X") 
ax[1].set_ylabel("1 Y")

ax[2].hist([1,2,3], [1,2,3], color= 'c')
ax[2].set_title("2 title") 
ax[2].set_xlabel("2 X") 
ax[2].set_ylabel("2 Y")

plt.show()
```

### 여러개의 그래프를 그릴 때- x축, y축 단위와 범위 바꾸기

```python
fig, ax = plt.subplots(1, 2, figsize=(20, 5)) 
# 0번 그래프
ax[0].plot([1,2,3], [1,2,3])
ax[0].set_xticks([1,2,3]) 

# 1번 그래프
ax[1].plot([1,2,3], [1,2,3]) 
ax[1].set_xlim([0, 5]) 

plt.show()
```

# 5. Seaborn
- Seaborn은 Matplotlib을 기반으로 다양한 색상 테마와 통계용 차트 등의 기능을 추가한 시각화 패키지이다.
- 더 이쁘게 그래프를 그릴 수 있다!!
- https://seaborn.pydata.org/api.html


```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
tips.head(6)
```

### Barplot

```python
ax = plt.plot()
ax = sns.barplot(x='time',y='total_bill',data=tips)
```

### Regplot
- 데이터를 점으로 나타내면서 선형성을 함께 확인한다.

```python
# 지불 금액에 따른 팁의 양 : ax 객체로 선언
ax = sns.regplot(x='total_bill',y='tip',data=tips ) 
# fit_reg=False 선형선 출력하지 않는 옵션
# ax = sns.regplot(x='total_bill',y='tip',
#                 data=tips, fit_reg=False ) 
ax.set_xlabel('TB')  # x축 이름 설정
ax.set_ylabel('Tip') # y축 이름 설정
ax.set_title('Total billl and Tip') # 그래프 제목 설정
```

### Jointplot
- 안쪽은 점 데이터로 분포를 확인하고 바깥쪽은 막대그래프로 밀집도를 확인한다.
- 데이터의 경향을 파악하기 좋다.

```python
joint = sns.jointplot(x='total_bill',y='tip',data=tips)

joint.set_axis_labels(xlabel='TB',ylabel='Tip')
# jointplot에서는 regplot과 라벨 붙이는 방식이 다름
# set_axis_labels 함수를 사용해서 라벨 설정 가능
# 굳이 라벨을 설정하지 않아도 기본 column명으로 라벨 형성
```

### Pairplot(data = tips)
- 수치에 해당하는 그래프를 전반적으로 그려줌
- 관계 그래프를 확인할 수 있음
- 전반적인 상태를 확인할 수 있어 처음 데이터를 확인할 때 전체를 파악하기 좋음

```python
sns.pairplot(data = tips)
```
