### +) 모든 요일의 웹툰정보를 크롤링하는 방법?

```python
week=['mon','tue','wed','thur','fri','sat','sun'] #요일 리스트
for i in week:
    url='https://comic.naver.com/webtoon/weekdayList.nhn?week={}'.format(i)
    print(url)
```

```python
week=['mon','tue','wed','thur','fri','sat','sun']
titles=[] #전체 요일의 제목 list
artists=[] #전체 요일의 작가 list
grades=[] #전체 요일의 평점 list
weeks=[] #요일을 담을 list
for i in week:
    wk=[i]
    url='https://comic.naver.com/webtoon/weekdayList.nhn?week={}'.format(i)
    response = requests.get(url) 
    html = response.text 
    soup = BeautifulSoup(html, "html.parser")
    title=[i.find('a').get('title') for i in soup.select('div.thumb')]
    artist=[i.find('a').get_text() for i in soup.select('dd.desc')]
    grade=[i.find('strong').get_text() for i in soup.select('div.rating_type')]
    title=title[3::]
    artist=artist[3::]
    titles=titles+title
    artists=artists+artist
    grades=grades+grade
    weeks=weeks+wk*len(title)
toon=pd.DataFrame({'제목':titles, '작가':artists, '평점':grades, '요일':weeks})
toon
```

## 1. Selenium
Selenium은 웹 테스팅 프레임워크이다.<br/> 웹 자동화 테스트에서 사용되고 또한 BeautifulSoup만으로는 스크래핑이 불가능할 때 사용할 수 있다.<br/> 예를 들어,
* 스크래핑 과정 중 웹 상에서의 동작이 필요할 때 ex) `.send_keys(), .click(), .move_to_element()`
* 페이지를 이리저리 돌아다녀도 주소가 변하지 않는 경우: 웹페이지가 `iframe`형식일때 ex) 네이버 카페
    * Selenium으로 iframe 전환 후 진행
    * 공동세션 자료 참고
* BeautifulSoup으로 잘한거 같은데 안될때...

Selenium은 웹페이지 내의 특정 요소(들)을 찾는 많은 메소드들을 제공하고 있는데, 이들은 보통 한 요소를 리턴하는 `find_element_---()` 혹은 복수 요소를 리턴하는 `find_elements_---()` 메소드로 구분된다. 웹페이지를 잘 보고 알맞게 사용하면 된다.

# Selenium에서 주로 활용하는 함수 

1. **get** : URL에 접속하게 해주는 함수
    * <font color = "green">ex )  driver.get('https://www.naver.com/') 


2. **find_element_by_xpath** : 해당 홈페이지에서 가져온 **Xpath**를 활용하여 원하는 값을 가져오게 해주는 함수
    * <font color = "green">ex )  driver.find_element_by_xpath('//*[@id="query"]')
    

3. **find_element_by_css_selector** : 홈페이지에서 가져온 **Copy selector**를 활용하여 원하는 값을 가져오게 해주는 함수 
    * <font color = "green">ex )  driver.find_element_by_css_selector('#account > div > a > i')
    

4. **click()** : xpath나 css_selector을 활용하여 불러온 위치를 클릭하게 해주는 함수
    * <font color = "green">ex )  driver.find_element_by_xpath('//*[@id="search_btn"]').click()
    

5. **send_keys()** : xpath나 css_selector을 활용하여 불러온 위치에 원하는 값을 입력해준다.  
    * <font color = "green">ex )  driver.find_element_by_xpath('//*[@id="query"]').send_keys('우한 폐렴')
    

6. **clear()** : xpath나 css_selector을 활용하여 불러온 위치에 입력한 값을 모두 지워준다. 
    * <font color = "green">ex )  driver.find_element_by_xpath('//*[@id="query"]').clear()


7. **time.sleep()** : ()안에 숫자를 입력하면 해당 시간(초)만큼 잠시 대기. 주로 셀레니움에서 홈페이지 로딩시간이나 스크롤을 내릴때 대기시간과 같이 페이지의 로딩이 필요할때나 데이터를 끌어올 때 서버에서 차단당하는 일을 최대한 막기위해 주로 사용한다.
    * <font color = "green">ex )  time.sleep(3)
    

## Selenium에서의 동작 정리:

|동작|Syntax|
|-----|-----|
|클릭|`element.click()`|

### xpath도 url처럼 규칙이 있습니다! 
/html/body/div[2]/div/div/main/div/div[3]/table/tbody/tr[1]/td[3]/a[1] <br/>
/html/body/div[2]/div/div/main/div/div[3]/table/tbody/tr[2]/td[3]/a[1] <br/>
/html/body/div[2]/div/div/main/div/div[3]/table/tbody/tr[3]/td[3]/a[1] <br/>
...<br/>
tr[]의 숫자가 바뀌는게 보이쥬? 

```python
# 데이터 저장할 list 설정
name=[]
start=[]
end=[]
length=[]
```

```python
for i in range(1,4): #1~3페이지만 크롤링합시다
    url='https://www.juso.go.kr/info/RoadNameDataList.do?currentPage={}&countPerPage=10&&city1=11&county1=590&town1=&keyword=&roadCd=&searchType=0&extend=true'.format(i)
    time.sleep(1)
    driver.get(url)
    time.sleep(1)
    for j in range(1,11): #각 도로명 클릭
        driver.find_element_by_xpath('/html/body/div[2]/div/div/main/div/div[3]/table/tbody/tr[{}]/td[3]/a[1]'.format(j)).click()
        time.sleep(2)
        html= driver.page_source
        soup= BeautifulSoup(html, 'lxml')
        name.append(soup.select_one('h2#title_addr_detail').get_text()) #도로 이름
        start.append(soup.select_one('td#SrtCn').get_text()) #기점
        end.append(soup.select_one('td#EndCn').get_text()) #종점
        length.append(soup.select_one('td#RoadLt').get_text()) #길이
        driver.get(url)
        time.sleep(2)
```

```python
df=pd.DataFrame({'도로명':name, '기점':start, '종점':end, '길이':length})
df
```
